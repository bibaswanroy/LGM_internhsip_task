<h2>Image to Pencil Sketch with Python</h2><br><br>
The task is to convert a normal RGB image to Pencil Sketch format

Importing the libraries

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = cv2.imread("cat.jpg")
# cv2.imshow("Cat", image)
# cv2.waitKey(0)

img = image[:,:,::-1]
plt.imshow(img)
```

2. Changing from RGB to Gray Scale

```python
grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grey, cmap="gray")
# cv2.imshow("CatNew", grey)
# cv2.waitKey(0)
```

3. Inverting the Gray Scale image

```python
inverted = 255 - grey
# cv2.imshow("Inverted",inverted)
# cv2.waitKey(0)
plt.imshow(inverted, cmap="gray")
```

4. Blurring the Grey Scale image

```python
blurred = cv2.GaussianBlur(inverted, (21, 21), 0)
# cv2.imshow("Blurred",blurred)
# cv2.waitKey(0)
plt.imshow(blurred, cmap="gray")
```

5. Reinverting to get a blurred Grey Scale in uninverted format

```python
blurred_inv = 255 - blurred
# cv2.imshow("BlurredInv",blurred_inv)
# cv2.waitKey(0)
plt.imshow(blurred_inv, cmap="gray")
```

6. Making the pencil sketch image

```python
pencil = cv2.divide(grey, blurred_inv, scale=256.0)
# cv2.imshow("Original",image)
# cv2.imshow("Pencil",pencil)
# cv2.waitKey(0)
plt.imshow(pencil, cmap="gray")
```
